```python
%matplotlib notebook

import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('C', 'E'), ('E', 'D'), ('D', 'G'), ('E', 'G'), ('G', 'F')])
nx.draw_networkx(G)
```

```python
nx.radius(G)
nx.diameter(G)
nx.nodes(G)
G.remove_node('G')
nx.is_connected(G)
T = nx.bfs_tree(G, 'A')
nx.draw_networkx(T)
nx.shortest_path_length(T,'A')
nx.center(G)
nx.periphery(G)
nx.eccentricity(G)
```

```python
%matplotlib notebook
import networkx as nx
import matplotlib.pyplot as plt
A1 = nx.DiGraph()
A1.add_edges_from([('A', 'C'), ('A', 'B'), ('B', 'C'), ('C', 'B'), ('D', 'C'), ('E', 'B'), ('C','D')])
nx.draw_networkx(A1)
B1 = nx.DiGraph()
B1.add_edges_from([('A', 'D'), ('D', 'E'), ('E', 'C'), ('B', 'C'), ('A', 'B'), ('B', 'A'), ('A', 'C'), ('C', 'A')])
nx.draw_networkx(B1)
C1 = nx.DiGraph()
C1.add_edges_from([('A', 'C'), ('B', 'A'), ('C', 'B'), ('A', 'D'), ('D', 'B')])
nx.draw_networkx(C1)
nx.is_weakly_connected(A1)
[c for c in nx.strongly_connected_components(B1)]
nx.is_strongly_connected(C1)
```
