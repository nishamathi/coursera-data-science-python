## Applied Social Network Analysis in Python - Week 3 Quiz

```python
%matplotlib notebook

import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('C', 'E'), ('E', 'D'), ('D', 'G'), ('E', 'G'), ('G', 'F')])
nx.draw_networkx(G)

degcentD = nx.degree_centrality(G)
degcentD['D']
closecentG = nx.closeness_centrality(G)
closecentG['G']
bwcentG = nx.betweenness_centrality(G, normalized=True, endpoints=False)
bwcentG['G']
bwcentedge = nx.edge_betweenness_centrality(G, normalized=False)
bwcentedge
```
